```python
import pandas as pd
from chemical_calculator import determine_organic_peroxide, ChemicalSpecies, ChemicalComposition, FormulatedProduct
```

# Chemical Calculator Example

```python
test_file_path = "../data/Product_List.csv"

df = pd.read_csv(test_file_path)

product_ids = sorted(df["Product ID"].value_counts().keys())

for product_id in product_ids:
    inds = df.index[df["Product ID"]==product_id]
    compositions = [ChemicalComposition(ChemicalSpecies(df["CAS"][ind], df["Chemical Name"][ind]), df["min (%)"][ind], df["max (%)"][ind]) for ind in inds]

    product = FormulatedProduct(product_id, compositions)
    is_organic_peroxide, reason = product.is_organic_peroxide()

    if is_organic_peroxide:
        print(f"{product_id} meets the definition of organic peroxide")
    else:
        print(f"{product_id} does not meet the definition of organic peroxide")

    print(f"Reason: {reason}")
```
